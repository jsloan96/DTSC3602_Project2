# Import Libraries

```
import pandas as pd
```

# Import Data

```
GTDB = pd.read_excel('globalterrorismdb.xlsx', header=0)
GTDB.head(10)
```

```
GTDB.shape
```

Filter for attacks in the United States.

```
GTDB_USA = GTDB[GTDB['country_txt'] == 'United States']
GTDB_USA
```

```
GTDB_USA = GTDB_USA.reset_index(drop=True)
GTDB_USA.head()
```

Export the new data set as a csv.

```
# saving the DataFrame as a CSV file
GTDB_USA_csv_data = GTDB_USA.to_csv('GTDB_USA.csv', index = False)
print('\nCSV String:\n', GTDB_USA_csv_data)
```
